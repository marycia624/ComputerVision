```
def exposion(image, p):
  new_img = np.round(image*(2**p), 0)
  new_img = new_img.astype(int)
  for i in range(new_img.shape[0]):
    for j in range(new_img.shape[1]):
      for k in range(new_img.shape[2]):
        if new_img[i][j][k] > 255:
          new_img[i][j][k] = 255
  return new_img

new_img = exposion(img, 0.46)
plt.imshow(new_img)
plt.show()
print(np.mean(new_img))
```

```
def contrast(image, p):
  new_img = image.copy()/255
  if p > 0:
    new_img = ((new_img-0.5)/(1-p))
  else:
    new_img = ((new_img - 0.5)*(1-p))
  
  for i in range(new_img.shape[0]):
    for j in range(new_img.shape[1]):
      for k in range(new_img.shape[2]):
        if new_img[i][j][k] > 1:
          new_img[i][j][k] = 1
        if new_img[i][j][k] < 0:
          new_img[i][j][k] = 0
  new_img = np.round(new_img*255)
  new_img = new_img.astype(int)
  return new_img

new_img = contrast(img, 0.46)
plt.imshow(new_img)
plt.show()
print(np.mean(new_img))
```

```
new_img = contrast(img, -0.22)
plt.imshow(new_img)
plt.show()
print(np.mean(new_img))
```

```

new_img = contrast(img, 0.99)
plt.imshow(new_img)
plt.show()
print(np.mean(new_img))
```
